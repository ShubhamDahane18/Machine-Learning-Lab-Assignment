```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set Seaborn style
sns.set(style='whitegrid')

# 1. Age distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Count plot of target variable (AHD)
plt.figure(figsize=(6, 4))
sns.countplot(x='AHD', data=df, palette='Set2')
plt.title('Heart Disease (AHD) Count')
plt.xlabel('AHD')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=df, palette='coolwarm')
plt.title('Gender Distribution')
plt.xticks([0, 1], ['Female', 'Male'])
plt.show()
```

```python
# 4. Boxplot - Cholesterol vs AHD
plt.figure(figsize=(8, 5))
sns.boxplot(x='AHD', y='Chol', data=df, palette='pastel')
plt.title('Cholesterol vs AHD')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Chol', hue='AHD', data=df, palette='husl')
plt.title('Age vs Cholesterol by Heart Disease')
plt.show()
```
